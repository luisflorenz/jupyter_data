## Exploratory Analysis and Visualization

Now we are going to create visualizations of the data frames to get a better idea of the data and to understand some of the variables that are representative in this data frame.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
import numpy as np
```

```python
countries_human_developed_df.describe()
```

```python
countries_human_developed_df.columns
```

```python
n_country = countries_human_developed_df[['Country','HDI rank','Life expectancy at birth', 'density'  ]]
n_country
```

### What is the correlation between HDI rank, Life expectancy at birth and population density?

```python
n_country.corr(method='pearson')
```

```python
plt.figure(figsize=(12,8))
plt.title('Correlation between HDI, Live expectancy and Dessity')
sns.heatmap(n_country.corr(method='pearson'), annot=True, cmap='Blues');
```

As we can see, the one that shows the most correlation is the life expectancy at birth with the population density.

### Is there any correlation between life expectancy at birth and the population growth rate?

```python
Live_Birth_Grownt_Population = countries_human_developed_df[['Life expectancy at birth', 'growthRate']]
Live_Birth_Grownt_Population
```

```python
Live_Birth_Grownt_Population_corr = Live_Birth_Grownt_Population.corr()
Live_Birth_Grownt_Population_corr
```

Here we see that there is a negative correlation between Life expectancy at birth and grown th rate.

Now we are going to plot it using the regplot method to see it more clearly.



```python
plt.figure(figsize=(12,8))
sns.regplot(x='Life expectancy at birth', y='growthRate', data=Live_Birth_Grownt_Population, scatter_kws={'s':100, 'alpha':0.5})
plt.title('Correlation of Life expectancy and growthRate ');
```

As we can see in this graph the correlation is negative, the regression line shows a negative slope. and as we can see, the data point that presents the greatest negative correlation data is life expectancy at birth between 70 and 75 years.

### Will the HDI rank, Life expectancy at birth, and the rank have any correlation?

```python
HDI_expectancy_rank = countries_human_developed_df[['Country', 'HDI rank','Life expectancy at birth', 'rank' ]]
HDI_expectancy_rank
```

```python
HDI_expectancy_rank.corr()
```

Let's see it now graphically to see the data more clearly.

```python
plt.figure(figsize=(12,8))
plt.title('Correlation HDI, Life expectancy at birth and Rank')
sns.heatmap(HDI_expectancy_rank.corr(), annot=True, cmap='YlGnBu');
```

As we can see, the one that shows the greatest correlation is life expectancy at birth with rank.
